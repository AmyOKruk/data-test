```python
import pandas as pd
```

```python
dams_df = pd.read_csv('associatedpress-us-dams/dams_in_nid_state_reports.csv')
```

```python
dams_df.head()
```

```python
dams_df.condition_assessment_standardized.value_counts()
```

```python
# Finding dams rated poor, unsatisfactory, poor_or_unsatisfactory, not_satisfactory 
```

```python
dams_poor = dams_df[(dams_df.condition_assessment_standardized == 'poor') | (dams_df.condition_assessment_standardized == 'unsatisfactory') | (dams_df.condition_assessment_standardized == 'poor_or_unsatisfactory') | (dams_df.condition_assessment_standardized == 'not_satisfactory')]
```

```python
dams_poor.head()
```

```python
dams_poor.hazard_standardized.value_counts()
```

```python
# Poor condition dams rated as high hazard
dang_dams_df = dams_poor[dams_poor.hazard_standardized == 'high']
```

```python
# Cleaning up the dataframe before I feed it to d3 to create an interactive map
```

```python
dang_dams_df.dam_name = dang_dams_df.dam_name.str.lower()
dang_dams_df.dam_name = dang_dams_df.dam_name.str.title()
```

```python
# Creating the final dataframe
```

```python
dang_dams_df2 = dang_dams_df[['dam_name', 'county', 'state', 'condition_assessment_standardized', 'hazard_standardized', 'last_inspection_date','year_dam_completed','latitude','longitude']]
```

```python
dang_dams_df2.county = dang_dams_df2.county + '_' + dang_dams_df2.state
```

```python
dang_dams_df2.shape
```

```python
dang_dams_df2['state2'] = dang_dams_df2.state.map(d)
```

```python
dang_dams_df2.year_dam_completed = dang_dams_df2.year_dam_completed.astype(int)
```
